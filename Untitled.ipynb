```python
import numpy as np
import pandas as pd
import regex as re
```

```python
ins = pd.read_csv('inscritos.csv'
                 ,header=1
                 ,sep=',')
```

```python
colunas=['col1','col2','col3','situacao','nro_inscricao','nome',
        'cpf','data_nascimento','genero','escolaridade',
        'instituicao','telefone','email','cep','bairro',
        'cidade','inclui_outro','perfil','observacao','perfis']
```

```python
ins.columns=colunas
```

```python
def columnsDrop():
    ins.drop('situacao', axis=1, inplace=True)
    ins.dropna(axis=1, how='all', inplace=True)
    ins.dropna(axis=0, how='all', inplace=True)

columnsDrop()
```

```python
ins.info()
```

```python
def columnsInsert():
    ins.insert(6, 'escolaridade_id',0)
    ins.insert(8, 'instituicao_id',0)
    ins.insert(14, 'cidade_id',0)
    ins.insert(18, 'tamanho_camiseta_id',0)

columnsInsert()
```

```python
def cidadeId(cidade):
    if cidade=='Guarulhos':
        return 1
    elif cidade=='São Paulo':
        return 2
    elif cidade=='Ribeirão Preto':
        return 3
    else:
        99

def tamanhoCamiseta(pTamanho) -> int:
    t=re.split('\s',pTamanho)
    if (t[len(t)-1]=='P'):
        return 1
    elif (t[len(t)-1]=='M'):
        return 2
    elif (t[len(t)-1]=='G'):
        return 3
    elif (t[len(t)-1]=='GG'):
        return 4
    else:
        return 5

ins['tamanho_camiseta_id']=ins['observacao'].apply(lambda x: tamanhoCamiseta(x))
ins['cidade_id']=ins['cidade'].apply(lambda x: cidadeId(x))
```
